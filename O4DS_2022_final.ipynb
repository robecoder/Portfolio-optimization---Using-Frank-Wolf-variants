```python
import numpy as np
import matplotlib.pyplot as plt
import time
import pandas as pd
#from yahoofinancials import YahooFinancials
from datetime import datetime
from timeit import default_timer
import seaborn as sns
import plotly.express as px
```

```python
def step_function(r, x_t, f_x, g,  d_t, x0, L):
  '''
  Adaptive Frank Wolfe method for SC function- Variant 2 stepsize 
  r   : data matrix 
  x_t : variable x at iteration k 
  f_x : value of obj function evaluated in the iterate x_k 
  g   : gap function of iterate x_k
  d_t : descent direction of iterate x_k
  x0  : previous iterate x_(k-1)
  L   : approximation of Lipschitz constant at the iteration k-1
    
  Returns
  gamma : stepsize for iteration k 
  M     : new value for Lipschitz constant for iteration k 
  '''

  if type(x0) == int:
    x0 = np.zeros(x_t.shape[0])
    
  epsilon = 10e-3 
  k1=0.9  
  k2 = 2
    
  #Lower bound for the approximation
  nL = k1 * L

  den=d_t.T @ d_t 
  if den==0: 
    den=1 
  den1=(2 * (f_x  - objective(x0, r)) * den)
  if den1==0:
    den1=1
    
  clip = g ** 2 / den1    
  M = np.clip(clip ,nL, L)         
    
  bound = g / (M * den) 
  gamma = min(bound, 1)

  while objective(x_t + (gamma * d_t), r) > (f_x - (gamma * g) + ( (gamma ** 2) * M) * (den)) : 
      M = k2 * M
      bound = g / ((M * den)+0.0000001)        
      gamma = min(bound, 1)
   
  return gamma,  M
```

The cost function is set such that we can deal with matrix-vector calculations:

*    $R\in\mathbb{R}^{T\times n}$, where $T$ is the number of periods and $n$ is the number of assets, so that the row $R[i,;]$ represents the return of each asset in the $i-th$ period;
*   $x\in\mathbb{R}^{n\times 1}$ that is the column vector that gives us the percentage to invest for each asset.

In this way the matrix multiplication between each row of $R$ and $x$ will correspond to exponential of the each element of the summatory $\sum_{i=1}^{T} ln(r_{i}x)$, where $T$ is the number of period. So what is left is to apply the logaritmic function for each term.

```python
def objective(x, r):
  x=x.reshape(-1,1) 
  a=np.dot(r, x )
  a=np.where(a!=0, a, a+0.0001) 
  
  #a.shape=(n_periods, 1)
  return np.sum(-np.log(a))
```

The gradient, whose dimension is:
\begin{equation}
\nabla f(x)=\left( \frac{\partial f}{\partial x_1},\ldots, \frac{\partial f}{\partial x_n} \right)\ \in\mathbb{R}^{1 \times n},
\end{equation}
has the following formulation:
\begin{equation}
\nabla f(x) =-\sum_{t=1}^{T} \frac{1}{r_t^T x}\ \overrightarrow{r}
\end{equation}
The computation of the gradient follows the same logic as before: we will get a column vector of $n$ components (i.e. the vector $x$) and then we will perform the dot product between this vector and the matrix $R$. Once this is done, we compute the inverse of this result and then element-wise multiply it for the vector of returns $r$. Finally, we do the sum for each period.

```python
def gradient(x, r):
  x=x.reshape(-1,1)
  a=np.dot(r, x )
  a=np.where(a!=0, a, a+0.0001) 
  
  return np.sum(-(1/a)*r, axis=0)
```

```python
def arimjo_ls(xt, f_x, gamma_max, gt, dt, r):
  '''
  Armijo Rule
  x_t : variable x at iteration k
  f_x : value of obj function evaluated in the iterate x_k
  gamma_max : maximum value for the gamma, it will eventually be shrinked
  gt : gap at iteration k
  dt : descent direction at iteration k
  r  : data matrix

  Returns
  gamma : stepsize 
  '''

  gamma = gamma_max
  p = 0.1*(-gt) #Â quantity that goes in the armijo line search
  f= objective(xt + gamma * dt, r)
  while (f > (f_x + gamma * p)): # while the function is biggr than the linear approximation
    gamma=0.7*gamma #reduce the step size
    f = objective(xt + gamma * dt, r) 
  return gamma  
```

```python
def LMO(gradient_xt): 
  '''
  Linear Minimizer Oracle (unit simplex)
  Returns
  st  : i_th vector of the basis of R^n
  sti : index that corresponds to the min of the gradient 
  '''

  sti = np.argmin(gradient_xt) 
  st = np.zeros(gradient_xt.shape[0])
  st[sti]=1 
  return st, sti
```

```python
def frank_wolfe(objective, gradient, return_matrix, max_iter):
    '''
    Classical Frank Wolfe (Unit Simplex)
    Returns
    x : solution to the investment distribution (|x| = n_assets)
    o   : list with objective function evaluations of each iteration  
    gap_list : list of gap function values of each iteration
    '''
    # n_periods = return_matrix.shape[0]
    n_assets = return_matrix.shape[1]
    
    eps = 1e-6    # Tolerance 
    gap_list = [] # Store gap function values 

    #Initial Approximation
    vertex_to_start = 1
    assert vertex_to_start < n_assets, print('Pick another initial approximation')
    x = np.zeros(n_assets)
    x[vertex_to_start] = 1  

    o = [objective(x, return_matrix)] # Store object function values

    algo_start = default_timer()
    iterations_counter = 0
    for k in range(1, max_iter + 1):
        iterations_counter += 1
       
        grad_vector = gradient(x, return_matrix)
        #assert(grad_vector.shape[0] == len(x))

        # FW for simplex problem --> minimum of linear  function will be reached 
        #                            on the vertex of the simplex
        # Here we search for index with minimum gradient component

        min_index = np.argmin(grad_vector)
        #assert grad_vector[min_index] <= 0, print(f'Minimum of gradient vector is positive')
        
        # x_k^ = e_ik from lectures - direction where to go
        xk_hat = np.zeros(n_assets)
        assert(len(xk_hat) == len(grad_vector))
        xk_hat[min_index] = 1
        d = (xk_hat - x)
        gap=-grad_vector @ d
        gap_list.append(gap)

        #Stopping Condition
        if gap <= eps:
            print(f'Satisfied exiting condition')
            break
        else:
            #Update
            learning_rate = 2 / (k + 1)
            x = x + learning_rate * d
            obj = objective(x, return_matrix)
            
            o.append(obj) 

    algo_end = default_timer()
    delta_t = algo_end - algo_start
    return x, o, gap_list, iterations_counter
```

```python
def variant2(objective,  gradient, r,  max_iter = 1000):
    ''' 
    Adaptive Frank Wolfe method for SC function - Variant 2 
    Returns a list with three elements
    x_t : final vector solution 
    o   : list with objective function evaluations of each iteration 
    gap_list : list of gap function values of each iteration
    '''

    eps = 1e-6    # Tolerance 
    gap_list = [] # Store gap function values 
    o = []        # Store object function values

    xt  = np.zeros(shape=r.shape[1])
    xt[0]=1
  
    #Keep Track of the previous value of x w.r.t. to the curret iteration
    x0=0 

    algo_start = default_timer()
    for i in range(max_iter):

      g  = gradient(xt, r)
      fxt  = objective(xt,r )
      o.append(fxt)

      # Linear Minimizer Oracle => st : descent direction point
      st, sti = LMO(g)
      d =  st - xt

      # Computing gap function
      gap = - g @ d 
      gap_list.append(gap)
      
      # Euristic formula => starting local Lipschitz constant L_{-1}
      if i==0:
        
        eps_l = 1e-3
        num = np.sqrt((g-gradient(xt + eps_l * d, r)) @ (g-gradient(xt + eps_l * d, r)))
        L = num / (eps_l * np.sqrt((d @ d)))

      #Stopping Condition  
      if gap<= eps: 
        print('early exit')
        algo_end = default_timer()
        delta_t = algo_end - algo_start
        return [xt, o, gap_list, i]
      
      gamma, L = step_function(r, xt, fxt, gap,  d, x0, L)
    
      # Updating values 
      x0 = xt         # new past value x_{k-1}
      xt += gamma*d   # new current value x_k

    algo_end = default_timer()
    delta_t = algo_end - algo_start

    return [xt,  o, gap_list, max_iter]
```

```python
def variant3(objective,  gradient, r,  max_iter = 1000):
  eps = 1e-6
  
      
  xt  = np.zeros(shape=r.shape[1])
  xt[0]=1
  
  x0=0 #initialize this variable to keep track of the previus value of x w.r.t. to the curret iteration
  #it is needed in the step_function
  gap_list=[]
  
  o=[] #store the objective function values during the iteration
  algo_start = default_timer()
  for i in range(max_iter):
    #if i==0:
        
    #    L=1 #initialize the L costant, to be sostituted with the right euristic
    
    g  = gradient(xt, r)
    fxt  = objective(xt,r )
    o.append(fxt)
    st, sti = LMO(g)#linear minimizer oracle
    d =  st - xt 
    gap=- g @ d #computing the gap
    gap_list.append(gap)
    if i==0:
      L = gap/(d@d)
      
    if gap<= eps: #checking the stopping condition
      print('early exit')
      algo_end = default_timer()
      delta_t = algo_end - algo_start
      return [xt, o, gap_list, i]
    if i == 0:
      gamma=1
    else:
      gamma, L = step_function(r, xt, fxt, gap,  d, x0, L)
    
    x0=xt #keeping track of the past value of x for the next iteration
    xt += gamma*d #updating x
  algo_end = default_timer()
  delta_t = algo_end - algo_start
  return [xt,  o, gap_list, max_iter]
```

```python
def FW_PW(objective, gradient, r, max_iter = 1000):
    '''
    Pairwise Frank Wolfe method 
    '''
    eps = 1e-6    # Tolerance 
    gap_list = [] # Store gap function values 
    o = []        # Store object function values 

    # List of indeces : verteces into set1 (see below)
    ind = [0] 

    # Create the starting point as a unit simplex vertex ei 
    xt = np.zeros(r.shape[1])
    xt[ind] = 1
    
    algo_start = default_timer()

    for i in range(max_iter):
        
        g = gradient(xt, r)
        fxt  = objective(xt, r)
        o.append(fxt)

        # Frank Wolfe descent point and index
        st, sti = LMO( g) # linear minimizer oracle
        
        # Away-step descent index and point 
        vti = np.argmax(g[ind])    # away vertex index 
        vti = ind[vti]             # away vertex index storing 
        
        vt = np.zeros(xt.shape[0]) 
        vt[vti] = 1                # away vertex e_{vti}

        dt  = st - vt              # pairwise direction
        gfw = -g @ dt              # gap
        gap_list.append(gfw)
        
        # Stopping condition \ Updating 
        if gfw <= eps: 
          print('early exit')
          algo_end = default_timer()
          delta_t = algo_end - algo_start
          return [xt, o, gap_list, i]
        
        gamma_max = xt[vti] # gamma max = the weight corresponding to the away vertex
        gamma = arimjo_ls(xt, fxt, gamma_max, gfw, dt, r)

        # Updating the list of indexes correspondig to the non zero weights
        if (abs(xt[vti]) <= 0.0001): # removing weights that are too little
            xt[vti] = 0

        ind = (np.array(list(range(xt.shape[0])))[(xt>0)]).tolist()
        
        xt += gamma*dt
    
    algo_end = default_timer()
    delta_t = algo_end - algo_start 

    return [xt, o, gap_list, max_iter]
```

```python
def plot_objective_function(values):
    fontsize = 20
    labelsize = 16
    fig, ax = plt.subplots()

    for a, s in values.items():
        ax.plot(s,  markerfacecolor = 'white', label=a)#marker = 'o',
        leg = ax.legend()
        #ax.set_yscale('log')
        ax.set_xscale('log')
    plt.xlabel('Number of iteration')
    plt.ylabel('Objective function')
    plt.title('Objective function')
    plt.grid()
```

```python
def plot_gap(values):
    fontsize = 20
    labelsize = 16
    fig, ax = plt.subplots()

    for a, s in values.items():
        ax.plot(s,  markerfacecolor = 'white', label=a) #marker = 'o',
        leg = ax.legend()
        ax.set_yscale('log')
        ax.set_xscale('log')
    plt.xlabel('Number of iteration')
    plt.ylabel('Gap')
    plt.title('Gap')
    plt.grid()
```

```python
def plot_return_matrix(return_matrix, x_lim, assets):
    fig, ax = plt.subplots()
    
    for asset in assets:
        ax.plot(return_matrix[:x_lim, asset], marker = 'o', markerfacecolor = 'white')

    plt.ylim([0, 1.5])
    plt.xlabel('Period 1 .. T')
    plt.title(f'First {x_lim} samples for {len(assets)} assets')
    plt.ylabel('Return')
    plt.legend([f'Asset {n}' for n in assets])
    plt.grid()
```

```python
def plot_bar_scores(scores_dict, y_label, title):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.set_ylabel(y_label, fontsize = 18)
    ax.set_title(title, fontsize = 18)

    x_vals = scores_dict.keys()
    y_vals = scores_dict.values()
    
    ax.bar(x_vals, y_vals, width = 0.4)
    ax.tick_params(labelsize = 16)
    plt.grid()
    plt.show()
```

```python
def run_experiment(n_assets, n_periods, max_iterations, fun, return_matrix=None):
    mean = 0 
    var = 0.1 
    np.random.seed(43)
    i={}
    o={}
    g={}
    t={}
    n_iterations = {}
    # Create return matrix as in the paper
    # (N periods) x (N assets) matrix
    if type(return_matrix)==type(None):
      return_matrix = np.ones((n_periods, n_assets)) + np.random.normal(loc = mean, scale = var, size = (n_periods, n_assets))
      assert(return_matrix.shape == (n_periods, n_assets))
      assert(len(return_matrix[:, 0]) == n_periods)
      assert(len(return_matrix[0, :]) == n_assets)
    


    plot_return_matrix(return_matrix, x_lim = 100, assets = [0, 1, 2])
    v=0
    for f in fun:
      
      name = f.__name__
      start=time.time()
      investment_distribution, objective_func, gap, n_iters = f(objective,  gradient, return_matrix,  max_iterations)
      stop=time.time()

      
      i[name]=investment_distribution
      o[name]=objective_func
      g[name]=gap
      t[name]=[stop-start]
      n_iterations[name] = n_iters

      if investment_distribution.sum()>=0.999:
        print(name, ' has done a correct execution')

    plot_objective_function(o)
    plot_gap(g)
    
    plot_bar_scores(
        n_iterations, 
        'Number of iterations', 
        'Iterations of algorithms'
    )

    plot_bar_scores(
        {k: v[0] for k, v in t.items()}, 
        'CPU time, sec', 
        'Computational time of algorithms'
    )

    invest_distributions = np.array([v for v in i.values()])
    print(invest_distributions.shape)
    #sns.heatmap(invest_distributions,  vmin=0, vmax=0.3)
    fig = px.imshow(invest_distributions, aspect="auto", y=[j for j in i.keys()])
    
    fig.show()
    display('Time to execution', pd.DataFrame(t))
    return i, o, g, t, n_iterations
```

```python
def plot_joint_barplot(list_of_dicts, title, ylab, xlab, xticklab, loc = 5, figsize = (6, 4)):
    d = pd.DataFrame.from_dict({k:[v] for k, v in list_of_dicts[0].items()})
    for i in range(1, len(list_of_dicts)):
        d = d.append({k:v for k, v in list_of_dicts[i].items()}, ignore_index = True)

    barplot = d.plot.bar(figsize = figsize)
    barplot.grid()

    fontsize = 15
    legendsize = 15

    barplot.tick_params(axis = 'both', labelsize = fontsize)
    barplot.set_title(title, fontsize = fontsize)
    barplot.set_ylabel(ylab, fontsize = fontsize)
    barplot.set_xlabel(xlab, fontsize = fontsize)
    barplot.set_xticklabels(xticklab, fontsize = fontsize)
    barplot.legend(prop = {'size': legendsize}, loc = loc)
        
    # d.head()

plot_joint_barplot(
    [n_iterations_800, n_iterations_1200, n_iterations_1500],
    title = 'Number of iterations for algorithms',
    xlab = 'Dataset size',
    ylab = 'Number of iterations',
    xticklab = [800, 1200, 1500]
)
```
